````r
library(ggplot2)
````

````r
participant <- c("A", "A", "B", "B", "C", "C", "D", "D")
time <- c(8.1, 3.4, 6.7, 4.2, 2.3, 5.2, 4.4, 5.5)
condition <- c("H", "C")
hfirst = c("H first", "H first", "C first", "C first")

df <- data.frame(participant, condition, time, hfirst)
df
````

````r
library(ggplot2)
labs_slope <- labs(
        title = "Stuff", 
        subtitle = "Human v computer gen",
        x = "condition", y = "reading time", 
        color = "Island")

ggp2_slope <- ggplot(data = df, 
    mapping = aes(x = condition,
                y = time, 
                group = participant)) +
  geom_line(aes(color = participant), 
                size = 2, show.legend = FALSE) +
  geom_point(aes(color = participant), 
                size = 4) +
  facet_grid(hfirst ~ .)

ggp2_slope + 
    labs_slope
````
